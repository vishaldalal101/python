```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#q = 1.602e-19
#c = 8.854e-12
m = 0.511017608


gamma = []
```

```python
def energylos_gamma(T,m):
            return (T/m)+1
        
for T in range(0,3000,10): 
    gamma.append(energylos_gamma(T,m))
    
#print(gamma)  
```

```python



plt.plot(gamma)
plt.show
    

plt.title('Graph for dependence of Etot to gamma')
plt.grid(color='brown', linestyle='-', linewidth=0.5)
plt.xlim(0, 3000)
plt.xlabel('Etot, [MeV]')
plt.ylim(0, 6500)
plt.ylabel('gamma')

plt.savefig('Graph_for_dependence_of_Etot_to_gamma.png')
```
